```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose
import pandas_datareader.data as web

!pip install yfinance --upgrade --no-cache-dir
import yfinance as yf
yf.pdr_override()
```

```python
itau= web.get_data_yahoo('ITUB4.SA')
itsa= web.get_data_yahoo('ITSA4.SA')
```

```python
itau.head()
itsa.head()
```

```python
itau_fatiado=itau[itau.index.year >= 2020]
itau_fatiado['Close'].plot(figsize=(16,8), label='ITUB4')
itau_fatiado['Close'].rolling(30).mean().plot(label='MM21')
itau_fatiado['Close'].rolling(200).mean().plot(label="MM200")
plt.legend()
plt.title('Grafico de Ações ITUB4')
plt.rcParams.update({'font.size': 25})
```

```python
itsa_fatiado=itsa[itsa.index.year >= 2020]
itsa_fatiado['Close'].plot(figsize=(16,8), label='ITUB4')
itsa_fatiado['Close'].rolling(21).mean().plot(label='MM21')
itsa_fatiado['Close'].rolling(200).mean().plot(label="MM200")
plt.legend()
plt.title('Grafico de Ações ITAUSA')
plt.rcParams.update({'font.size': 25})
```
