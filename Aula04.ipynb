```
def renda_maxima (x):
   if x.Q006 == "A": return(0)
   elif x.Q006 == "B": return(998)
   elif x.Q006 == "C": return(1497)
   elif x.Q006 == "D": return(1996)
   elif x.Q006 == "E": return(2495)
   elif x.Q006 == "F": return(2994)
   elif x.Q006 == "G": return(3992)
   elif x.Q006 == "H": return(4990)
   elif x.Q006 == "I": return(5988)
   elif x.Q006 == "J": return(6986)
   elif x.Q006 == "K": return(7984)
   elif x.Q006 == "L": return(8982)
   elif x.Q006 == "M": return(9980)
   elif x.Q006 == "N": return(11976)
   elif x.Q006 == "O": return(14976)
   elif x.Q006 == "P": return(14970)
   elif x.Q006 == "Q": return(19960)
   else: return(20000)

dados['RENDA_MAXIMA'] = dados[['Q006']].apply(renda_maxima, axis=1)

#Q025Na sua residência tem internet? A = Não; B = Sim, sem notas zero
def internet (x):
   if x.Q025 == "A": return("Não")
   elif x.Q025 == "B": return("Sim")
   else: return(null)

dados['TEM_INTERNET'] = dados[['Q025']].apply(internet, axis=1)

#Q024Na sua residência tem computador? A = Não; B a E = Sim, sem notas zero
def computador (x):
   if x.Q024 == "A": return("Não")
   elif x.Q024 == "B" or x.Q024 == "C" or x.Q024 == "D" or x.Q024 == "E" : return("Sim")
   else: return(null)

dados['TEM_COMPUTADOR'] = dados[['Q024']].apply(computador, axis=1)



#Q005 Incluindo você, quantas pessoas moram atualmente em sua residência?
def renda_max_percapta (x):
  salario_minimo = 998
  renda_percapta = x.RENDA_MAXIMA/x.Q005
  salarios_minimos_porpessoa = int(renda_percapta/salario_minimo)
  return(salarios_minimos_porpessoa)

dados['SALARIOS_MINIMOS_PERCAPTA'] = dados[['Q005','RENDA_MAXIMA']].apply(renda_max_percapta, axis=1)

dados.head()

dados_sem_notas_zero = dados.query("NU_NOTA_TOTAL != 0") #retirando inscritos que faltaram em todos os dias, ou tirou zero em todas as provas

dados_sem_notas_zero.head()

dados_sem_nota_zero_rendapercapta_salariominimo = dados.query("SALARIOS_MINIMOS_PERCAPTA >= 1") #retirando inscritos com renda por habitante na casa inferior ao salário mínimo da epoca

dados_sem_nota_zero_rendapercapta_salariominimo.head()
```

```
def funcaoBoxSplotRenda(nota):
  renda_Maxima_Ordenada = dados['RENDA_MAXIMA'].unique()
  renda_Maxima_Ordenada.sort()
  plt.figure(figsize=(15,6))
  title = nota + " pela Renda (Renda Máxima da Faixa)"
  plt.title(title)
  sns.boxplot(x="RENDA_MAXIMA", y=nota, data=dados, hue='TEM_COMPUTADOR', order=renda_Maxima_Ordenada)

provas = ['NU_NOTA_CN', 'NU_NOTA_CH', 'NU_NOTA_MT', 'NU_NOTA_LC', 'NU_NOTA_REDACAO']
dados['NU_NOTA_TOTAL'] = dados[provas].sum(axis=1)
provas.append('NU_NOTA_TOTAL')

for i in provas:
  funcaoBoxSplotRenda(i)
```

```
x = notas_entrada
y = notas_saida 
```

```
from sklearn.model_selection import train_test_split

SEED = 4321

x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size = 0.25,
                                                        random_state=SEED)
```

```
from sklearn.svm import LinearSVR

modelo = LinearSVR(random_state = SEED)
modelo.fit(x_treino, y_treino)
```

```
predicoes_matematica = modelo.predict(x_teste)
```

```
plt.figure(figsize=(8, 8))
sns.scatterplot(x=predicoes_matematica, y=y_teste)
plt.xlim((-50, 1050))
plt.ylim((-50, 1050))
```

```
plt.figure(figsize=(8, 8))
sns.scatterplot(x=y_teste, y=x_teste.mean(axis=1))
plt.xlim((-50, 1050))
plt.ylim((-50, 1050))
```

```
resultados = pd.DataFrame()
resultados["Real"] = y_teste
resultados["Previsao"] = predicoes_matematica
resultados["diferenca"] = resultados["Real"] - resultados["Previsao"]
resultados["quadrado_diferenca"] = (resultados["Real"] - resultados["Previsao"])**2
resultados
```
